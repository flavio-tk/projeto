```
from pyspark.sql.functions import lit, col, trim, when, count
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType
from datetime import datetime, timedelta
from delta.tables import DeltaTable
```

```
current_date = datetime.now()

# Calcula o início do mês anterior
previous_month = current_date.replace(day=1) - timedelta(days=1)
previous_year_month = previous_month.strftime("%Y%m")
year = int(previous_month.strftime("%Y"))
month = int(previous_month.strftime("%m"))
begin_previous_month = previous_month.replace(day=1).strftime("%Y%m%d")

# Calcula o fim do mês anterior
end_previous_month = previous_month.strftime("%Y%m%d")

print(previous_year_month)
print(year)
print(month)
```

```
df_carga_mensal =  spark.table("bronze.nasa_solar_flux_temperature").filter((col("YEAR") == year) & (col("MO") == lit(month))) 
```

```
df_carga_mensal_tranformacao =  df_carga_mensal \
     .select(col("YEAR"),
             col("MO").cast('Integer').alias("MO"),
             col("DY").cast('Integer').alias("DY"),
             (when((trim(col('ALLSKY_SFC_SW_DWN')) == lit('-999.0')) | (col('ALLSKY_SFC_SW_DWN').isNull()) , 0)
              .otherwise(trim(col('ALLSKY_SFC_SW_DWN')))).cast('float').alias('ALLSKY_SFC_SW_DWN'),
             (when((trim(col('CLRSKY_SFC_SW_DWN')) == lit('-999.0')) | (col('CLRSKY_SFC_SW_DWN').isNull()), 0)
              .otherwise(trim(col('CLRSKY_SFC_SW_DWN')))).cast('float').alias('CLRSKY_SFC_SW_DWN'),
             (when((trim(col('ALLSKY_KT')) == lit('-999.0')) | (col('ALLSKY_KT').isNull()), 0)
              .otherwise(trim(col('ALLSKY_KT')))).cast('float').alias('ALLSKY_KT'),
             (when((trim(col('ALLSKY_SFC_LW_DWN')) == lit('-999.0')) | (col('ALLSKY_SFC_LW_DWN').isNull()), 0)
              .otherwise(trim(col('ALLSKY_SFC_LW_DWN')))).cast('float').alias('ALLSKY_SFC_LW_DWN'),
             (when((trim(col('ALLSKY_SFC_PAR_TOT')) == lit('-999.0')) | (col('ALLSKY_SFC_PAR_TOT').isNull()), 0)
              .otherwise(trim(col('ALLSKY_SFC_PAR_TOT')))).cast('float').alias('ALLSKY_SFC_PAR_TOT'),
             (when((trim(col('CLRSKY_SFC_PAR_TOT')) == lit('-999.0')) | (col('CLRSKY_SFC_PAR_TOT').isNull()), 0)
              .otherwise(trim(col('CLRSKY_SFC_PAR_TOT')))).cast('float').alias('CLRSKY_SFC_PAR_TOT'),
             (when((trim(col('ALLSKY_SFC_UVA')) == lit('-999.0')) | (col('ALLSKY_SFC_UVA').isNull()), 0)
              .otherwise(trim(col('ALLSKY_SFC_UVA')))).cast('float').alias('ALLSKY_SFC_UVA'),
             (when((trim(col('ALLSKY_SFC_UVB')) == lit('-999.0')) | (col('ALLSKY_SFC_UVB').isNull()), 0)
              .otherwise(trim(col('ALLSKY_SFC_UVB')))).cast('float').alias('ALLSKY_SFC_UVB'),
             (when((trim(col('ALLSKY_SFC_UV_INDEX')) == lit('-999.0')) | (col('ALLSKY_SFC_UV_INDEX').isNull()), 0)
              .otherwise(trim(col('ALLSKY_SFC_UV_INDEX')))).cast('float').alias('ALLSKY_SFC_UV_INDEX'),
             (when((trim(col('T2M')) == lit('-999.0')) | (col('T2M').isNull()), 0)
              .otherwise(trim(col('T2M')))).cast('float').alias('T2M'),
             (when((trim(col('T2MDEW')) == lit('-999.0')) | (col('T2MDEW').isNull()), 0)
              .otherwise(trim(col('T2MDEW')))).cast('float').alias('T2MDEW'),
             (when((trim(col('T2MWET')) == lit('-999.0')) | (col('T2MWET').isNull()), 0)
              .otherwise(trim(col('T2MWET')))).cast('float').alias('T2MWET'),
             (when((trim(col('TS')) == lit('-999.0')) | (col('TS').isNull()), 0)
              .otherwise(trim(col('TS')))).cast('float').alias('TS'),
             (when((trim(col('T2M_RANGE')) == lit('-999.0')) | (col('T2M_RANGE').isNull()), 0)
              .otherwise(trim(col('T2M_RANGE')))).cast('float').alias('T2M_RANGE'),
             (when((trim(col('T2M_MAX')) == lit('-999.0')) | (col('T2M_MAX').isNull()), 0)
              .otherwise(trim(col('T2M_MAX')))).cast('float').alias('T2M_MAX'),
             (when((trim(col('T2M_MIN')) == lit('-999.0')) | (col('T2M_MIN').isNull()), 0)
              .otherwise(trim(col('T2M_MIN')))).cast('float').alias('T2M_MIN')
            )
# df_carga_mensal_tranformacao.display()
```

```
df_estacao_semana = spark.table("silver.dim_data_estacao_semana")

df_carga_mensal_tranformacao_2 = df_carga_mensal_tranformacao.alias("cmt") \
          .join(df_estacao_semana.alias("es"), (col("cmt.YEAR") == col("es.YEAR")) & 
                                               (col("cmt.MO") == col("es.MO")) &
                                               (col("cmt.DY") == col("es.DY")) , "left") \
          .select(col("cmt.YEAR"),
                  col("cmt.MO"),
                  col("cmt.DY"),
                  col("es.SEASON_CODE"),
                  col("es.WEEK_CODE"),
                  col("cmt.ALLSKY_SFC_SW_DWN"),
                  col("cmt.CLRSKY_SFC_SW_DWN"),
                  col("cmt.ALLSKY_KT"),
                  col("cmt.ALLSKY_SFC_LW_DWN"),
                  col("cmt.ALLSKY_SFC_PAR_TOT"),
                  col("cmt.CLRSKY_SFC_PAR_TOT"),
                  col("cmt.ALLSKY_SFC_UVA"),
                  col("cmt.ALLSKY_SFC_UVB"),
                  col("cmt.ALLSKY_SFC_UV_INDEX"),
                  col("cmt.T2M"),
                  col("cmt.T2MDEW"),
                  col("cmt.T2MWET"),
                  col("cmt.TS"),
                  col("cmt.T2M_RANGE"),
                  col("cmt.T2M_MAX"),
                  col("cmt.T2M_MIN")
                 )
# df_carga_mensal_tranformacao_2.display()
```

```
delta_table_path = "/mnt/projeto_climatico/silver/nasa_solar_flux_temperature"
condicao_exclusao = f"YEAR = {year} AND MO = {month}"
delta_table = DeltaTable.forPath(spark, delta_table_path)
delta_table.delete(condicao_exclusao)
```

```
df_carga_mensal_tranformacao_2.write.mode("append").option("path", "/mnt/projeto_climatico/silver/nasa_solar_flux_temperature").format("delta").saveAsTable("silver.nasa_solar_flux_temperature")
```
